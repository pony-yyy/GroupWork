```python
# Import the necessary libraries
from google.cloud import storage
from google.cloud import bigquery

project_id = "bdao-ima-422711"
dataset_name = "bobbdaoima"
table_name = "restaurant_checkin" # if you make changes previously, then here you need to make according change

bigquery_client = bigquery.Client(project=project_id)
```

```python
# Query all data from the table
query = f"""
   SELECT DISTINCT business_id, total_number_of_checkin
   FROM `{project_id}.{dataset_name}.{table_name}`
   ORDER BY business_id;

"""

query_job = bigquery_client.query(query)
results = query_job.result()

checkin_df = query_job.to_dataframe()
checkin_df
```

```python
from google.cloud import bigquery

project_id = "bdao-ima-422711"

dataset_name = "bobbdaoima"
table_name = "checkin_data"

bigquery_client = bigquery.Client(project=project_id)

job_config = bigquery.LoadJobConfig(
    write_disposition="WRITE_TRUNCATE",
)

job = bigquery_client.load_table_from_dataframe(
    checkin_df,
    f"{project_id}.{dataset_name}.{table_name}",
    job_config=job_config,
)

job.result()

print(f"Data uploaded to BigQuery table {dataset_name}.{table_name}")
```
