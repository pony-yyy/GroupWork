```python
import pandas as pd

def select_top_categories_with_percentages(df, selected_cols, num_top_categories=10):
    # 确定不在 selected_columns 中的列
    category_cols = [col for col in df.columns if col not in selected_cols]

    # 计算每个种类的 1 的数量
    category_counts = df[category_cols].sum().sort_values(ascending=False)

    # 选择出现次数最多的 num_top_categories 个种类
    top_categories = category_counts.head(num_top_categories).index

    # 创建一个新的 DataFrame，只包括 top_categories
    top_categories_df = df[top_categories]

    # 将所有其他种类归类到 'others' 中
    df['others'] = df[category_cols].drop(columns=top_categories).sum(axis=1)

    # 计算每个种类的占比
    total_counts = df[category_cols].sum().sum()
    category_percentages = (category_counts / total_counts) * 100  # 计算百分比

    # 将 selected_columns 与 top_categories_df 合并
    result_df = pd.concat([df[selected_cols], top_categories_df, df[['others']]], axis=1)

    return result_df, category_percentages.head(num_top_categories)  # 返回 DataFrame 和占比

# 假设 restaurant_df 是原始的 DataFrame
# 假设 selected_columns 是已选择的列名列表
result_df, top_categories_percentages = select_top_categories_with_percentages(restaurant_df, selected_columns.columns)

# 查看处理后的 DataFrame 结构和种类占比
print(result_df.head())
print(top_categories_percentages)
```

```python
def calculate_hours(open_hours):
    if open_hours == '' or open_hours is None:
        return 0
    if open_hours == '0:0-0:0':
        return 24
    open_time, close_time = open_hours.split('-')
    open_hour, open_minute = map(int, open_time.split(':'))
    close_hour, close_minute = map(int, close_time.split(':'))

    if close_hour < open_hour or (close_hour == open_hour and close_minute < open_minute):
        close_hour += 24

    return (close_hour + close_minute / 60) - (open_hour + open_minute / 60)
```

```python
import ast
result_df['hours'] = result_df['hours'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for day in days_of_week:
    # Apply the calculate_hours function only if x is a dictionary
    result_df[day + '_hours'] = result_df['hours'].apply(lambda x: calculate_hours(x.get(day, '')) if isinstance(x, dict) else 0)

# Calculate total weekly hours by summing daily hours
result_df['Total_Weekly_Hours'] = result_df[[day + '_hours' for day in days_of_week]].sum(axis=1)

# Count the number of days per week that have more than 0 hours of operation
result_df['Operating_Days_per_Week'] = result_df[[day + '_hours' for day in days_of_week]].apply(lambda x: (x > 0).sum(), axis=1)

# Determine if there is operation on Sunday by checking if Sunday hours are greater than 0
result_df['Sunday_Operating'] = (result_df['Sunday_hours'] > 0).astype(int)
```
